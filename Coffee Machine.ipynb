```python
class CoffeeMachine:
    
    water = 400   
    milk = 540
    beans = 120
    cups = 9
    money = 550
    
    def __init__(self):
    
        self.state = "" # keeps track of state of machine (buy, fill, take, remaining, exit)
        self.ing = 0 # keeps track which ingredient getting filled
    
    #action() deals with user input
    def action(self, Input):
        if self.state == "":
            if Input == 'buy':
                self.state = 'buy'
                print("What do you want to buy? 1 - espresso, 2 - latte, 3 - cappuccino, back - to main menu:")
            elif Input == 'fill':
                self.state = 'fill'
                print("Write how many ml of water do you want to add:")
            elif Input == 'take':
                self.state = 'take'
                self.take()
                print("Write action (buy, fill, take, remaining, exit):")
            elif Input == 'remaining':
                self.state = 'remaining'
                self.remaining()
                print("Write action (buy, fill, take, remaining, exit):")
            elif Input == 'exit':
                return 1
            else:
                print("Write action (buy, fill, take, remaining, exit):")
            
        else:
            if self.state == 'buy':
                choice = Input
                self.buy(choice)
                print("Write action (buy, fill, take, remaining, exit):")
            elif self.state == 'fill':
                if self.ing == 0:
                    CoffeeMachine.water += int(Input)
                    self.ing += 1
                    print("Write how many ml of milk do you want to add:")
                elif self.ing == 1:
                    CoffeeMachine.milk += int(Input)
                    self.ing += 1
                    print("Write how many grams of coffee beans do you want to add:")
                elif self.ing == 2:
                    CoffeeMachine.beans += int(Input)
                    self.ing += 1
                    print("Write how many disposable cups of coffee do you want to add:")
                else:
                    CoffeeMachine.cups += int(Input)
                    self.ing = 0
                    self.state = ""
                    print("Write action (buy, fill, take, remaining, exit):")
        return 
    
    # check if there's enough ingredients to make the chosen coffee
    def check(self, choice):
    
        flags = {'water': 0,
                 'milk': 0,
                 'beans': 0,
                 'cups': 0}
        flag = 0
        if (CoffeeMachine.water < 250 and choice == '1') or (CoffeeMachine.water < 350 and choice == '2') or (CoffeeMachine.water < 200 and choice == '3'):
            flags['water'] = 1
        
        if (CoffeeMachine.beans < 16 and choice == '1') or (CoffeeMachine.beans < 20 and choice == '2') or (CoffeeMachine.beans < 12 and choice == '3'):
            flags['beans'] = 1
        
        if (CoffeeMachine.milk < 75 and choice == '2') or (CoffeeMachine.milk < 100 and choice == '3'):
            flags['milk'] = 1
        
        if CoffeeMachine.cups < 1:
            flags['cups'] = 1
        
        shortage = []
        for key in flags.keys():
            if flags[key] == 1:
                flag = 1
                shortage.append(key)
        if flag == 1:
            self.state = ""
            string = "Sorry, not enough "
            short_string = ', '.join(shortage)        
            print(string+short_string+'!')
        return flag
    
    
    def buy(self, choice):
        
        if choice == '1':
            if not self.check(choice):
                CoffeeMachine.water -= 250
                CoffeeMachine.beans -= 16
                CoffeeMachine.cups -= 1
                CoffeeMachine.money += 4
                print("I have enough resources, making you a coffee!")
                
            self.state = ""
        elif choice == '2':
            if not self.check(choice):
                CoffeeMachine.water -= 350
                CoffeeMachine.milk -= 75
                CoffeeMachine.beans -= 20
                CoffeeMachine.cups -= 1
                CoffeeMachine.money += 7
                print("I have enough resources, making you a coffee!")
                
            self.state = ""
        elif choice == '3':
            if not self.check(choice):
                CoffeeMachine.water -= 200
                CoffeeMachine.milk -= 100
                CoffeeMachine.beans -= 12
                CoffeeMachine.cups -= 1
                CoffeeMachine.money += 6
                print("I have enough resources, making you a coffee!")
            self.state = ""
        else:
            self.state = ""
            return
        
        
        
    def take(self):
        print("I gave you ${}".format(CoffeeMachine.money))
        CoffeeMachine.money = 0
        self.state = ""
    
    def remaining(self):
        print("The coffee machine has:")
        print(CoffeeMachine.water,"of water")
        print(CoffeeMachine.milk,"of milk")
        print(CoffeeMachine.beans,"of coffee beans")
        print(CoffeeMachine.cups,"of disposable cups")
        print("${} of money".format(CoffeeMachine.money))
        self.state = ""
```

```python
coffee = CoffeeMachine()
print("Write action (buy, fill, take, remaining, exit):") 
x = 0
while x == 0:
    action = input()
    if coffee.action(action):
         x = 1
```
